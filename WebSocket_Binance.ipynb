```python
import websocket, json
from datetime import datetime
#from google.colab import output


```

```python
instrument="bnbusdt"
interval="1m"
```

```python
socket = f"wss://stream.binance.com:9443/ws/{instrument}@kline_{interval}"
```

```python
socket2 = "wss://btc.data.hxro.io/live"
```

```python
import json
from google.colab import output




def on_message(ws, message):
    jsoned = json.loads(message)
    #print(jsoned)
    time = (jsoned["E"])
    time = datetime.fromtimestamp(int(str(time)[:10])).strftime('%M:%S')
    currency = (jsoned["s"])
    price = (jsoned["k"]['c'])
    openPrice = (jsoned["k"]['o'])
    marketMove = ""
    if (openPrice < price):
      marketMove = "Up"
    elif (openPrice > price):
      marketMove = "Down"
    else:
      marketMove = "Equilibrium"

    print(f'Time: {time} | Currency Symbol: {currency} | Start Price {openPrice} | Current Price {price} | Market movement: {marketMove}')
    
   

def on_close(ws, close_status_code, close_msg):
    print("Connection closed now")

ws = websocket.WebSocketApp(socket,on_message = on_message, on_close = on_close)
ws.run_forever()
```

```python
def on_message2(ws, message):
    jsoned = json.loads(message)
    time = (jsoned["ts"])
    time = datetime.fromtimestamp(int(str(time)[:10])).strftime('%M:%S')
    price = jsoned["price"]
    print(f'Time: {time} | Price: {price}')

def on_close2(ws, close_status_code, close_msg):
    print("Connection closed now")

    

ws = websocket.WebSocketApp(socket2, on_message = on_message2, on_close = on_close2)

ws.run_forever()
```
